```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
print(df)
print(df.columns)

dfs = df.sort_values(by='A', ascending=True)

dfs.plot(x='A', y=['B', 'C'])

plt.show()

dfs.plot(x='A', y=['B', 'C'], logx=True, logy=True)

plt.show()
```

# Αναφορά
    Από την πρώτη γραφική παράσταση, δεν μπορούμε να βγάλουμε συμπεράσματα.
    Από την δεύτερη γραφική παράσταση ωστόσο, που παίρνουμε το γράφημα του Β και του C ως προς τον Α σε λογαριθμική κλίμακα, βγάζουμε συμπεράσματα και για τις δυο σχέσεις.
    Για την σχέση του Α με τον Β καταλαβαίνουμε ότι είναι ΠΟΛΥΩΝΥΜΙΚΉ, επειδή η δεύτερη γραφική παράσταση μας δίνει αποτέλεσμα γραμμικής σχέσης.
    Για την σχέση του Α με τον C καταλαβαίνουμε ότι είναι ΕΚΘΕΤΙΚΉ, επειδή η δεύτερη γραφική παράσταση μας δίνει αποτέλεσμα εκθετικής σχέσης.
