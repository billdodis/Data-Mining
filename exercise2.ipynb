```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
print(df)
print(df.columns)

dfs = df.sort_values(by='A', ascending=True)

dfs.plot(x='A', y=['B', 'C'])

plt.show()

dfs.plot(x='A', y=['B', 'C'], logx=True, logy=True)

plt.show()
```
