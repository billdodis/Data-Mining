```python
import random

def sample(filename, k):
    sampleArray = []
    counter = 0
    with open(filename) as file:
        for line in file:
            counter += 1
            line = line.rstrip()
            if counter <= k:
                sampleArray.append(line)
            j = random.randrange(counter)
            if j < k:
                sampleArray[j] = line
        print(sampleArray)
        return
# THIS PYTHON FILE HAS TO BE IN THE SAME FOLDER AS THE INPUT.TXT FILE IN ORDER TO RUN CORRECTLY!
```

# Απόδειξη
    Χωρίζουμε την απόδειξη σε τρεις περιπτώσεις.
    
    1) Το αντικείμενο ανήκει στα πρώτα Κ αντικείμενα:
        P(εμφάνισης στην τελική εκτύπωση) = P(αποθήκευσης αντικειμένου) * P(παραμονής μέχρι το τέλος του ρεύματος)
        P(αποθήκευσης αντικειμένου) = 1 , γιατί τα πρώτα Κ αντικείμενα τα αποθηκεύουμε χωρίς κριτήριο
        P(παραμονής μέχρι το τέλος του ρεύματος) = ?
        Για να παραμείνει μέχρι το τέλος του ρεύματος, πρέπει να μην αντικατασταθεί από κανένα αντικείμενο που διατρέχεται από τον αλγόριθμο. Μπορεί να μπει ένα αντικείμενο στον χώρο αντικειμένων που κρατάει ο αλγόριθμος, αλλά όχι στην δική του θέση.
        Έτσι έχουμε:
        Συμβολίζω με χ = 'δεν εισέρχεται στην θέση του αντικειμένου'.
        P(παραμονής μέχρι το τέλος του ρεύματος) = P(χ ο Κ+1) * P(χ ο Κ+2) * ... * P(χ ο Ν-1) * P(χ ο Ν)
        P(χ ο Κ+1) = Κ/Κ+1 , γιατί μόνο σε μια περίπτωση παίρνει την θέση του αντικειμένου μας, δηλαδή 1/Κ+1, Κ+1 είναι όλα τα πιθανά αποτελέσματα του τυχαίου αριθμού για την επιλογή του Κ+1 αντικειμένου. 
        P(o K+1 εισέρχεται στην θέση του αντικειμένου) = 1/K+1
        P(o K+1 δεν εισέρχεται στην θέση του αντικειμένου) = 1 - P(o K+1 εισέρχεται στην θέση του αντικειμένου) = Κ/Κ+1
        Με την ίδια λογική έχω:
        P(χ ο Κ+2) = Κ+1/Κ+2
        P(χ ο Ν-1) = Ν-2/Ν-1
        P(χ ο Ν) = Ν-1/Ν   και θα έχω τους ίδιους τύπους για τις πιθανότητες των αντικειμένων [Κ+3, Ν-2].
        P(παραμονής μέχρι το τέλος του ρεύματος) = Κ/Κ+1 * Κ+1/Κ+2 * ... * Ν-2/Ν-1 * Ν-1/Ν
        Όλοι οι παρονομαστές φεύγουν με τον αριθμητή του επόμενου αντικειμένου και καταλήγω σε:
        P(παραμονής μέχρι το τέλος του ρεύματος) = Κ/Ν
        P(εμφάνισης στην τελική εκτύπωση) = 1 * K/N = K/N
        
    2) Το Κ+1 αντικείμενο:
        P(εμφάνισης στην τελική εκτύπωση) = P(αποθήκευσης αντικειμένου) * P(παραμονής μέχρι το τέλος του ρεύματος)
        P(αποθήκευσης αντικειμένου) = Κ/Κ+1 , γιατί μόνο στην περίπτωση που τύχει Κ+1 ο τυχαίος αριθμός δεν αποθηκεύεται το             αντικείμενο.
        Συμβολίζω με χ = 'δεν εισέρχεται στην θέση του αντικειμένου'.
        P(παραμονής μέχρι το τέλος του ρεύματος) = P(χ ο Κ+2) * P(χ ο Κ+3) * ... * P(χ ο Ν-1) * P(χ ο Ν)
        P(παραμονής μέχρι το τέλος του ρεύματος) = Κ+1/Κ+2 * Κ+2/Κ+3 * ... * Ν-2/Ν-1 * Ν-1/Ν
        Όλοι οι παρονομαστές φεύγουν με τον αριθμητή του επόμενου αντικειμένου και καταλήγω σε:
        P(παραμονής μέχρι το τέλος του ρεύματος) = Κ+1/Ν
        P(εμφάνισης στην τελική εκτύπωση) = K/K+1 * K+1/N = K/N
        
    3) Το Ν αντικείμενο (ΤΕΛΕΥΤΑΙΟ):
        P(εμφάνισης στην τελική εκτύπωση) = P(αποθήκευσης αντικειμένου) , γιατί είναι το τελευταίο αντικείμενο του ρεύματος.
        P(αποθήκευσης αντικειμένου) = Κ/Ν , γιατί μόνο στην περίπτωση που τύχει [0,Κ-1] (Κ αριθμοί) ο τυχαίος αριθμός αποθηκεύεται το αντικείμενο.
        
    Σε όλες τις περιπτώσεις η πιθανότητα εμφάνισης του αντικειμένου στην τελική εκτύπωση είναι Κ/Ν.
